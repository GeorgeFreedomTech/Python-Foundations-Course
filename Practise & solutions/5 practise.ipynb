````python
# Solutions for Practise I

# 1. Data Log Analysis
sensor_data = [10, 25, 8, 42, 15, 30]

print(f"Minimum reading: {min(sensor_data)}")
print(f"Maximum reading: {max(sensor_data)}")
print(f"Number of readings: {len(sensor_data)}")
print(f"Average reading: {sum(sensor_data) / len(sensor_data)}") 

even_readings = []
odd_readings = []

for reading in sensor_data:
    if reading % 2 == 0:
        even_readings.append(reading)
    else:
        odd_readings.append(reading)

print(f"Even readings: {even_readings}. Odd readings: {odd_readings}.")

# --

# 2. Challenge: Data Log Analysis
data_values = [10, 25, 8, 42, 15, 30]

# Manual minimum
minimum_val = data_values[0]
for val in data_values:
    if val < minimum_val:
        minimum_val = val
print(f"Manual Minimum: {minimum_val}")

# Manual maximum
maximum_val = data_values[0]
for val in data_values:
    if val > maximum_val:
        maximum_val = val
print(f"Manual Maximum: {maximum_val}")

# Manual length (item count)
item_count = 0
for _ in data_values:
    item_count += 1
print(f"Manual Item Count: {item_count}")

# Manual average (assumes item_count will be > 0)
total_sum = 0
for val in data_values:
    total_sum += val
average_val = total_sum / item_count
print(f"Manual Average: {average_val}")

# Even/Odd numbers
manual_even = []
manual_odd = []
for val in data_values:
    if val % 2 == 0:
        manual_even.append(val)
    else:
        manual_odd.append(val)
print(f"Manual Evens: {manual_even}. Manual Odds: {manual_odd}.")

# --

# 3. Dynamic Data Acquisition & Averaging
data_points_list = []
num_entries_str = input("How many data points for this survey? ")
num_entries = int(num_entries_str) # Assumes valid integer input

for i in range(num_entries):
    point_str = input(f"Enter data point #{i + 1}: ")
    data_points_list.append(int(point_str)) # Assumes valid integer input

average_data = sum(data_points_list) / len(data_points_list)
print(f"The average of the recorded data points is: {average_data}")

# --

# 4. Log Entry Validation (Timestamps)
validated_timestamps = []
CURRENT_EXPEDITION_YEAR = 2025
MIN_VALID_YEAR = 1800 # Example lower bound

num_years_str = input(f"How many log timestamps (years) to enter (e.g., 4-5)? ")
num_years = int(num_years_str) # Assuming valid integer input

for i in range(num_years):
    year_str = input(f"Enter timestamp year #{i + 1}: ")
    entry_year = int(year_str)
    
    if entry_year > MIN_VALID_YEAR and entry_year <= CURRENT_EXPEDITION_YEAR:
        validated_timestamps.append(entry_year)
            
print(f"Validated timestamps: {validated_timestamps}")

# --

# 5. Explorer's Inventory Management (Basic)
inventory = []
print("Equipping for the new expedition...") # Thematic print
for i in range(3): # Add 3 items
    item = input(f"Add item #{i + 1} to inventory (e.g., Compass, Rope, Medkit): ")
    inventory.append(item)

print("\nCurrent Inventory:") # Thematic print
for item in inventory:
    print(item) # Simple print, one item per line

item_to_remove = input("Enter item name to discard from inventory: ")
inventory.remove(item_to_remove) 

print("\nUpdated Inventory:") # Thematic print
for item in inventory:
    print(item)

# --

# 6. Challenge: Interactive Inventory System
explorer_backpack = [] # Using a different name for clarity

while True:
    print("\n--- Expedition Backpack Menu ---") # Thematic title
    print("1. Add item to backpack")
    print("2. View backpack contents")
    print("3. Discard item from backpack")
    print("4. End expedition (Quit)")
    
    choice = input("Choose action (1-4): ")

    if choice == "1":
        item_to_add = input("What item to add to backpack? ")
        explorer_backpack.append(item_to_add)
        print(f"'{item_to_add}' added to backpack.")

    elif choice == "2":
        if not explorer_backpack:
            print("Backpack is empty.")
        else:
            print("\nBackpack Contents:")
            for item in explorer_backpack:
                print(item)

    elif choice == "3":
        if not explorer_backpack:
            print("Backpack is empty, nothing to discard.")
        else:
            item_to_discard = input("What item to discard from backpack? ")
            if item_to_discard in explorer_backpack:
                explorer_backpack.remove(item_to_discard)
                print(f"'{item_to_discard}' discarded from backpack.")
            else:
                print(f"'{item_to_discard}' not found in backpack.")

    elif choice == "4":
        print("Ending expedition and closing inventory system. Safe travels!") # Thematic exit
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 4.")



# Solutions for Practise II 

# 1. Mission Resource Management
mission_supplies = {
    'Medkit': {'value': 25, 'category': 'Medical', 'quantity': 10},
    'Power Cell': {'value': 50, 'category': 'Energy', 'quantity': 25},
    'Protein Bar': {'value': 5, 'category': 'Food', 'quantity': 100},
    'Navigation Chart': {'value': 150, 'category': 'Navigation', 'quantity': 3},
    'Signal Flare': {'value': 0, 'category': 'Emergency', 'quantity': 50}
}

operative_gear = {
    'Agent Alpha': ['Medkit', 'Power Cell', 'Signal Flare'],
    'Dr. Aris': ['Navigation Chart', 'Protein Bar'],
    'Engineer Jian': ['Power Cell', 'Protein Bar', 'Signal Flare', 'Medkit'],
    'Scout Eva': ['Medkit', 'Protein Bar']
}

# Calculate total value of all supplies in stock
total_stock_value = 0
for item_name in mission_supplies: # Iterates through keys of the dictionary
    if mission_supplies[item_name]['quantity'] > 0:
        total_stock_value += mission_supplies[item_name]['value'] * mission_supplies[item_name]['quantity']
print(f"Total value of supplies in stock: {total_stock_value}")

# Create a list of all operatives carrying a 'Medkit'
medkit_carriers = []
for operative_name in operative_gear: # Iterates through keys (operative names)
    if 'Medkit' in operative_gear[operative_name]: # Checks if 'Medkit' is in the list of gear for that operative
        medkit_carriers.append(operative_name)
print(f"Operatives carrying 'Medkit': {medkit_carriers}")

# Find the supply item with the lowest value (excluding 0)
priced_items_list = []
for item_name in mission_supplies:
    if mission_supplies[item_name]['value'] > 0:
        priced_items_list.append([item_name, mission_supplies[item_name]['value']])

# Find the minimum among these (assuming priced_items_list is not empty after filtering)
if priced_items_list: # Check if there are any priced items
    lowest_value_item = priced_items_list[0] # Initialize with the first item
    for item_info in priced_items_list:
        if item_info[1] < lowest_value_item[1]: # Compare values (at index 1)
            lowest_value_item = item_info
    print(f"Lowest value item (not free): {lowest_value_item[0]} with value {lowest_value_item[1]}")
else:
    print("No priced items found (excluding free items).")

# --

# 2. Challenge: Resource Analysis
# Using mission_supplies and operative_gear from Exercise 1

# Count unique supply items carried by all operatives
unique_carried_items = set()
for operative_name in operative_gear: # Iterate through operative names (keys)
    for item_name in operative_gear[operative_name]: # Iterate through list of items for each operative
        unique_carried_items.add(item_name)
print(f"Number of unique items carried by operatives: {len(unique_carried_items)}")
print(f"Unique items carried: {unique_carried_items}")

# Create a new dictionary: keys are supply category, values are lists of item names
items_by_category = {}
for item_name in mission_supplies: # Iterate through item names (keys) in supplies
    category = mission_supplies[item_name]['category']
    if category not in items_by_category: # If category is a new key
        items_by_category[category] = [] # Initialize an empty list for this category
    items_by_category[category].append(item_name) # Add item name to the list for its category
print(f"Supplies organised by category: {items_by_category}")

# --

# 3. Data Anomaly Report Processing
anomaly_report_data = {
    'report_generated': True,
    'title': "Sector Gamma Anomaly Scan",
    'scan_points': 1024,
    'avg_signal_strength': 75.34,
    'data_sequence_gen': (x*x for x in range(3)), # A generator object
    'unique_regions_scanned': {1, 2, 3, 2, 1},
    'sensor_log_summary': ['OK', 'OK', 'ALERT'],
    'metadata_dict': {'source': 'Probe7', 'priority': 'High'},
    'enumerated_waypoints': enumerate(['WaypointAlpha', 'WaypointBeta']), # An enumerate object
    'fixed_params_tuple': (100, 200),
    'scan_range_obj': range(5), # A range object
    'critical_codes_fs': frozenset(['ALPHA', 'OMEGA']),
    'raw_signal_bytes': b'SignalData'
}

# Iterate through values and print their data type
print("\nData types in anomaly_report_data:")
for value_item in anomaly_report_data.values():
    print(type(value_item))

# Store all keys in a list and print
key_list = list(anomaly_report_data.keys())
print(f"\nKeys from report: {key_list}")

# Convert key list to set, compare lengths
key_set = set(key_list)
# Printing using a multi-line f-string as in the CZ solution example
print(f"""
    Number of items in key list: {len(key_list)}
    Number of items in key set: {len(key_set)}
""")
# An additional check to confirm if counts are same (as per EN exercise text)
if len(key_list) == len(key_set):
    print("List and Set of keys have the same number of elements (as expected, keys are unique).")
else:
    print("List and Set of keys have different numbers of elements (this indicates an issue).")
````
